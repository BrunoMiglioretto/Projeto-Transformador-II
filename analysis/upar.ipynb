```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
train_examples_path = "/Users/brunomiglioretto/puc/upar_challenge/data/annotations/phase1/train/train.csv"

df = pd.read_csv(train_examples_path)
```

```python
df.shape
```

```python
df.describe()
```

```python
counts = []

for column in df.columns:
    a = df[column].value_counts()
    print(a)
    counts.append([column, "0", a[0], abs(a[0] - a[1])])
    counts.append([column, "1", a[1], abs(a[0] - a[1])])

counts_df = pd.DataFrame(counts, columns=["Feature", "Class", "Count", "Difference"])
counts_df = counts_df.sort_values("Difference")
counts_df
```

```python
g = sns.catplot(
    data=counts_df,
    kind="bar",
    x="Feature",
    y="Count",
    hue="Class",
    height=4,
    aspect=4
)

g.set_xticklabels(rotation=90)

plt.show()
```
